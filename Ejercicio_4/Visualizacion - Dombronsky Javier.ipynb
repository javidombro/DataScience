```python
#1. Cargar este archivo en Python. Realizar estadísticas descriptivas básicas
import numpy
import pandas

hr_data = pandas.read_csv('HRDataset_v14.csv')
hr_data.groupby(by='Sex')['Salary'].describe()
```

```python
#2. Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?
import matplotlib
import seaborn

seaborn.histplot(hr_data.Salary)
print('Los salarios mas populares se encuentran en el rango 50000 - 100000')
#no estoy seguro si esto habia que calcularlo o simplemente como hice ahora viendo el grafico..
```

```python
#3. Realizar un grafico violin con los salarios discriminados por genero.

seaborn.violinplot(x='Salary', y='Sex', title='Salarios discriminados por genero', data=hr_data)
```

```python
#4. Graficar la serie de tiempo correspondiente a la fecha de contratacion
x = pandas.to_datetime(hr_data.DateofHire)
x.index = pandas.DatetimeIndex(x)
data = x.resample('Y').count()
ax = seaborn.lineplot(x= data.index, y=data.values, data=data)
ax.set(xlabel='Año',ylabel='Contrataciones', title='Contrataciones por año')
```
